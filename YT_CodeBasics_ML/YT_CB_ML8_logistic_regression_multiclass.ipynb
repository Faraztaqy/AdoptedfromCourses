In this tutorial we will see how to use logistic regression for multiclass classification. 

```python
from sklearn.datasets import load_digits
# digit dataset
# 8x8 1797 images
%matplotlib inline
import matplotlib.pyplot as plt
digits = load_digits() # trainin set 
```

<h4 style='color:purple'>Create and train logistic regression model</h4>

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, train_size=0.8)
print(len(X_train))
print(len(X_test))
```

```python
model.fit(X_train, y_train) 
# x_train will have the image of the handwritten
# y_train will have the target
```

<h4 style='color:purple'>Measure accuracy of our model</h4>

```python
model.score(X_test, y_test)
```

```python
#plt.matshow(digits.images[0:5])
for i in range(4):
    plt.matshow(digits.images[i])
print(digits.target[0:5]) # actual dataset
print(model.predict(digits.data[0:5])) # predict from dataset
```

<h4 style='color:purple'>Confusion Matrix</h4>

 to know which dataset that fail

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
# y_test: truth
# y_predicted: what model predicted
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
# top-left: no. of times the model predicted as well as the truth for that digit number
# for example no. 8, means the model predicted as different number than the truth being fed
# anywhere in the black area where it is not zero, means the model is not working properly
```
